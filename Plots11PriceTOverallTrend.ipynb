```python
# 4 room histogram for Price VS Distance from MRT


%matplotlib inline 
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import parseUnitDetails

# To see trend of 4-Room flats available for Nov 2018 Sale of balance flats by probability vs prices
unitDetails =parseUnitDetails.getUnitDetails()
labels = ['Ang Mo Kio','Bedok','Bukit Batok','Bukit Merah','Bukit Panjang','Choa Chu Kang',
                'Clementi','Geylang','Hougang','Jurong East','Jurong West','Kallang / Whampoa','Pasir Ris','Punggol',
                'Queenstown','Sembawang','Sengkang','Serangoon','Tampines','Toa Payoh', 'Woodlands','Yishun']
# defining names
DistToMrt= unitDetails['DistToMrt']
town = unitDetails['town']
rmType= unitDetails['rmType']
sellingPrice=unitDetails['sellingPrice']




# Gathering data for town, distance and price
distPrice=list(zip(town,rmType,DistToMrt,sellingPrice))


dist_town=[]
price_town=[]
for name in unitDetails:
    if name['rmType']=='4-Room':
        dist_town.append(name['DistToMrt'])  # not that townToMrt is not a array
        price_town.append(name['sellingPrice'])
    price=[float(i) for i in price_town]


    
# the histogram of the 4-Room Flat
fig = plt.figure()
fig.set_size_inches(18.5, 10.5, forward=True)
n, bins, patches = plt.hist(price, 100, density=True, facecolor='lightsalmon', alpha=0.75, edgecolor='black', linewidth =1.0)
plt.text(330000, .0000085, 'First Peak:\nHypothesize is Non Mature estate', fontsize=15,color='b')
plt.text(550000, .000006, 'Second Peak:\nHypothesize is Mature estate', fontsize=15,color='b')

plt.xlabel('prices',fontsize=18)

plt.ylabel('Probability',fontsize=18)
plt.title('Histogram of Price fluctuation for 4-Room flats',fontsize=22)
plt.grid(True)
plt.show()
```
