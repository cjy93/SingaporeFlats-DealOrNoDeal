```python
# Selling price over level for 4-Room flats
%matplotlib inline 
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import os
from mpl_toolkits.mplot3d import Axes3D

os.getcwd()
# TO get a boxplot to see the number of 4-Rooms and number of 5-Room flats available for Nov 2018 Sale of balance flats

unitDetails = np.genfromtxt("Merged_unitsDetails_DistToMrt.csv", delimiter=",",skip_header=1,
                dtype=[('town','U24'),('rmType','U7'),('blk','U6'),('unitNum','U10'),
                      ('level','i4'),('sqm','i8'),
                      ('sellingPrice','i8'),('color','U12') ,('repurchasedFlat','U30'),('leaseLessThan60Yrs','U10'),
                      ('street','U30'),('probableCompletionDate','U10'),('deliveryPossessionDate','U30'),
                      ('leaseCommencementDate','U15'),('availableEthnicQuotaMl','U20'),
                      ('availableEthnicQuotaCh','U25'),('availableEthnicQuotaIn','U20'),('lastUpdate','U60'),
                      ('NearestMrt','U30'),('DistToMrt','f8'),('blkX','i8'),('blkY','i8')],
                missing_values=['na','-',''],filling_values=0, comments='!')

labels = ['Ang Mo Kio','Bedok','Bukit Batok','Bukit Merah','Bukit Panjang','Choa Chu Kang',
                'Clementi','Geylang','Hougang','Jurong East','Jurong West','Kallang / Whampoa','Pasir Ris','Punggol',
                'Queenstown','Sembawang','Sengkang','Serangoon','Tampines','Toa Payoh', 'Woodlands','Yishun']
# defining names
town = unitDetails['town']
rmType= unitDetails['rmType']
floorlevel = unitDetails['level']
priceLevel =unitDetails['sellingPrice']
#print(town)



##############################################

# Gathering data for town, distance and price

countPerLevel=list(zip(town,rmType,floorlevel,priceLevel))

# make subplots from my above plots
# We can ask for ALL THE AXES and put them into axes

# Subplots  
## bars for floor level
fig, axes = plt.subplots(nrows=5, ncols=5, sharex=False, sharey=True, figsize=(15,15))
axes_list = [item for sublist in axes for item in sublist] 
  


for town in labels: 
    floorlevel_town=[]    
    pricelevel_town=[]
    levelNum_town =[]
    for name in countPerLevel:
        if name[0]== town and name[1]=='4-Room':
            floorlevel_town.append(name[2]) 
            pricelevel_town.append(name[3])
     
    
    # zipping them up floor and price per level 
    priceVsLevel = dict(zip(floorlevel_town,pricelevel_town))
    #print(priceVsLevel)  
   
    
    
    # making floors in consequtive order for graph x axis
    if len(floorlevel_town) > 0:
        maxFloor = max(floorlevel_town) # max of empty list is error so split to these cases
       
    else:
        maxFloor = 1 # dummy so levelNum dont have to loop so much
      
    #print(maxFloor)
    #print('maxPrice\n',maxPrice)    
    levelNum = list(range(1,maxFloor+1))
    
    # z= count number of units per level 
    countUnitsPerLevel = [floorlevel_town.count(i) for i in floorlevel_town]    

    
    
  

    ax = axes_list.pop(0)
    xs,ys=floorlevel_town, pricelevel_town
    
    #ax.bar(xs,ys,
    #z=countUnitsPerLevel,zdir='z',color='mediumpurple', alpha=0.75,projection='3d')
    ax.bar(xs, ys, color='mediumpurple', alpha=0.6)
 
        
 
    ax.set_title(town)
    ax.tick_params(
        which='both',
        bottom=False,
        left=False,
        right=False,
        top=False
    )
    ax.grid(linewidth=0.25)
    #ax.set_xlim((1950, 2020))
    #ax.set_xticks(range(0,2800, 1000))
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.grid(linewidth=0.25)
    ax.yaxis.grid(True)
    #ax.set_xlim((0,3000))
    ax.set_xlabel("Level")
    ax.set_ylabel("Selling Price")
    fig.suptitle('4-Room Flats Price of units per floor', fontsize =20)
    

    
for ax in axes_list:
    ax.remove() # remove extra empty subplots


plt.tight_layout()  
fig.subplots_adjust(top=0.91)
plt.show()    
print()
    
        
## Conclusion
# some purple is darker due to overlaying (multiple same level per town)
# The common intuition is that lower floors are cheaper than higher floors, but it seems we have little to prove that here.
```
