```python
# Get the number of units per level per town for 4-Room
%matplotlib inline 
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import os


os.getcwd()
# To get a boxplot to see the number of 4-Room flats available for Nov 2018 Sale of balance flats

unitDetails = np.genfromtxt("Merged_unitsDetails_DistToMrt.csv", delimiter=",",skip_header=1,
                dtype=[('town','U24'),('rmType','U7'),('blk','U6'),('unitNum','U10'),
                      ('level','i4'),('sqm','i8'),
                      ('sellingPrice','i8'),('color','U12') ,('repurchasedFlat','U30'),('leaseLessThan60Yrs','U10'),
                      ('street','U30'),('probableCompletionDate','U10'),('deliveryPossessionDate','U30'),
                      ('leaseCommencementDate','U15'),('availableEthnicQuotaMl','U20'),
                      ('availableEthnicQuotaCh','U25'),('availableEthnicQuotaIn','U20'),('lastUpdate','U60'),
                      ('NearestMrt','U30'),('DistToMrt','f8'),('blkX','i8'),('blkY','i8')],
                missing_values=['na','-',''],filling_values=0, comments='!')

labels = ['Ang Mo Kio','Bedok','Bukit Batok','Bukit Merah','Bukit Panjang','Choa Chu Kang',
                'Clementi','Geylang','Hougang','Jurong East','Jurong West','Kallang / Whampoa','Pasir Ris','Punggol',
                'Queenstown','Sembawang','Sengkang','Serangoon','Tampines','Toa Payoh', 'Woodlands','Yishun']
# defining names
town = unitDetails['town']
rmType= unitDetails['rmType']
floorlevel = unitDetails['level']
#print(town)



##############################################

# Gathering data for town, distance and price

countPerLevel=list(zip(town,rmType,floorlevel))

# make subplots from my above plots
# We can ask for ALL THE AXES and put them into axes
fig, axes = plt.subplots(nrows=3, ncols=8, sharex=False, sharey=True, figsize=(15,15))
axes_list = [item for sublist in axes for item in sublist] 

for town in labels: 
    floorlevel_town=[]    
    for name in countPerLevel:
        if name[0]== town and name[1]=='4-Room':
            floorlevel_town.append(name[2]) 
    if len(floorlevel_town) > 0:
        maxFloor = max(floorlevel_town) # max of empty list is error so split to these cases
    else:
        maxFloor = 1 # dummy so levelNum dont have to loop so much
        
    #print(maxFloor)
    levelNum = list(range(1,maxFloor+1))

    # count number of units per level 
    countLevel = [floorlevel_town.count(i) for i in levelNum]
    #print(countLevel)
                
   
    # bars for floor level
    ax = axes_list.pop(0)
    ax.bar(levelNum, countLevel, facecolor='g', alpha=0.75)

 
    
    ax.set_title(town)
    ax.tick_params(
        which='both',
        bottom=False,
        left=False,
        right=False,
        top=False
    )
    ax.grid(linewidth=0.25)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.grid(linewidth=0.25)
    #ax.set_xlim((0,3000))
    ax.set_xlabel("Level")
    ax.set_ylabel("Number of Units")
    fig.suptitle('4-Room Flats count of number of units per floor',fontsize=20)
    

    
for ax in axes_list:
        ax.remove()

plt.tight_layout()  
fig.subplots_adjust(top=0.91)
plt.show()    
print()
    
        
## Conclusion
# Typically , Low floors have more sale of balance units than higher floors
```
