```python
# rate of subscription for 4 room Flats
%matplotlib inline 
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import os
import matplotlib.patches as mpatches

os.getcwd()
########################################################
# 4-Room flats

# import data on Application rate
estatesMaturity  = np.genfromtxt("Number of Applications Received for 3-room and bigger flats as at 19 Nov 2018.csv", delimiter=",",skip_header=1,
                dtype=[('town','U24'),('FlatType','U50'),('No_of_Units','i8'),('Number_of_applicants','i8'),
                      ('Rate_Non_Elderly_First_timers','f8'),('Rate_Non_Elderly_Second_timers','f8'),
                      ('Rate_Non_Elderly_Overall','f8'),('Mature_Estates','U8')],
                missing_values=['na','-','','NA'],filling_values=0)

labels = np.array(['Ang Mo Kio','Bedok','Bukit Batok','Bukit Merah','Bukit Panjang','Choa Chu Kang',
                'Clementi','Geylang','Hougang','Jurong East','Jurong West','Kallang / Whampoa','Pasir Ris','Punggol',
                'Queenstown','Sembawang','Sengkang','Serangoon','Tampines','Toa Payoh', 'Woodlands','Yishun'])

estatesMaturity_subset_rmType = estatesMaturity[estatesMaturity['FlatType'] == '4-room'] # caused pasir ris to be gone because pasir ris no 4 room flat

estatesMaturityTown= estatesMaturity_subset_rmType['town']
estatesMaturityMature= estatesMaturity_subset_rmType['Mature_Estates']
overallRate= estatesMaturity_subset_rmType['Rate_Non_Elderly_Overall']



estatesMaturityCon= zip(estatesMaturityTown,estatesMaturityMature,overallRate)

# make unique sets of 'towns' and 'Mature Estates'
estatesMaturityCon =set(estatesMaturityCon)


nMature = []
yMature = []

print(estatesMaturityCon)
for town in estatesMaturityCon:
    if town[1]=='N':
        nMature.append(town[0])
    elif town[1]=='Y':
        yMature.append(town[0])

yMature = list(set(yMature))
nMature = list(set(nMature))
yMature.sort()
nMature.sort()
#print("Mature estates : \n{}".format(yMature))
#print("Non Mature estates : \n{}".format(nMature))

Mature = []
for repeatMature in range(len(yMature)):
    Mature.append('Mature')


NonMature = []
for repeatMature in range(len(yMature)):
    NonMature.append('Non Mature')


## Making labels in order
label = yMature + nMature

# Make different colour for Mature and Non Mature estates  
values= Mature + NonMature
keys = label
#print('len k,v', len(keys), len(values))

# COlour based on mature or non mature
# change fill color
region_color_dict = {
    "Mature": "pink",
    "Non Mature": "lavender"
}
town_region_dict= dict(zip(keys,values))  #### dict used this way diff

colourlist = [region_color_dict[town_region_dict[town]] for town in estatesMaturityTown]
#print(colourlist)
  
# plotting bar graph
fig = plt.figure(figsize=(15,4))
ax = fig.add_subplot(111)
bargraph = plt.bar(estatesMaturityTown,overallRate, color=colourlist)
ax.text(len(estatesMaturityTown), 1, 'rate = 1', style='italic',
        bbox={'facecolor':'skyblue', 'alpha':0.5, 'pad':10})
linegraph = plt.hlines(1,-1,len(estatesMaturityTown)+1,'b')
plt.xticks(fontsize=11,rotation='vertical')
plt.ylabel('Application Rate')
plt.title('Application rate for 4-Room Flats')


# Add Labels using mpatches library
pink_patch = mpatches.Patch(color='pink', label='Mature')
lavender_patch = mpatches.Patch(color='lavender', label='Non Mature')
plt.legend(handles=[pink_patch,lavender_patch])



plt.show()    
print()
    
```
