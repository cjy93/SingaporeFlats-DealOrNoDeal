```python
# Price Vs level for 4-room flat according to count of number of flats per level
%matplotlib inline 
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import os
import parseUnitDetails
from mpl_toolkits.mplot3d import Axes3D

os.getcwd()
# TO get a boxplot to see the number of 4-Room flats available for Nov 2018 Sale of balance flats
# bring in csv data file for unit details
unitDetails = parseUnitDetails.getUnitDetails()

labels = ['Ang Mo Kio','Bedok','Bukit Batok','Bukit Merah','Bukit Panjang','Choa Chu Kang',
                'Clementi','Geylang','Hougang','Jurong East','Jurong West','Kallang / Whampoa','Pasir Ris','Punggol',
                'Queenstown','Sembawang','Sengkang','Serangoon','Tampines','Toa Payoh', 'Woodlands','Yishun']
# defining names
town = unitDetails['town']
rmType= unitDetails['rmType']
floorlevel = unitDetails['level']
priceLevel =unitDetails['sellingPrice']
#print(town)



##############################################

# Gathering data for town, distance and price

countPerLevel=list(zip(town,rmType,floorlevel,priceLevel))

# make subplots from my above plots
# We can ask for ALL THE AXES and put them into axes
fig, axes = plt.subplots(nrows=3, ncols=8, sharex=False, sharey=True, figsize=(18,12), subplot_kw={'projection':'3d'})
axes_list = [item for sublist in axes for item in sublist] 

for town in labels: 
    floorlevel_town=[]    
    pricelevel_town=[]
    levelNum_town =[]
    for name in countPerLevel:
        if name[0]== town and name[1]=='4-Room':
            floorlevel_town.append(name[2]) 
            pricelevel_town.append(name[3])
      
    
    # zipping them up floor and price per level 
    priceVsLevel = dict(zip(floorlevel_town,pricelevel_town))
    #print(priceVsLevel)  
   
    
    
    # making floors in consequtive order for graph x axis
    if len(floorlevel_town) > 0:
        maxFloor = max(floorlevel_town) # max of empty list is error so split to these cases
       
    else:
        maxFloor = 1 # dummy so levelNum dont have to loop so much
      
    #print(maxFloor)
    #print('maxPrice\n',maxPrice)    
    levelNum = list(range(1,maxFloor+1))
    
    # z= count number of units per level 
    countUnitsPerLevel = [floorlevel_town.count(i) for i in floorlevel_town]    

    
    
  
    # Subplots       
    ax = axes_list.pop(0)
    xs,ys=floorlevel_town, countUnitsPerLevel
    zs=pricelevel_town
    #print('x: ', floorlevel_town, ' y: ', countUnitsPerLevel, ' z: ', pricelevel_town)
    #ax.bar(xs,ys,
    #z=countUnitsPerLevel,zdir='z',color='mediumpurple', alpha=0.75,projection='3d')
    ax.bar(xs, zs, ys , zdir='y', color='mediumpurple', alpha=0.8)
 
        
    
    #ax.plot(dist_town_sorted, price_town_sorted ,'.',clip_on=False)
    #ax.plot([0,1,2] )
    ax.set_title(town)
  
    fig.suptitle('4-Room Flats Price of units per floor',fontsize =20)
    ax.set_xlabel('Level')
    ax.set_ylabel('Count per level')
    ax.set_zlabel('Selling Price')
    

    
for ax in axes_list:
    ax.remove() # remove extra empty subplots

plt.show()    
print()
    
        
## Conclusion
# some purple is darker due to overlaying (multiple same level per town)
# The common intuition is that lower floors are cheaper than higher floors, but it seems we have little to prove that here.
```
